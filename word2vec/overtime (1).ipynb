```python
"""author: sarah konrad, data+ 2023
this is an updated and adapted version of overtime.ipynb created by the 2022 cohort.
this version includes a function to track euclidean distances between words overtime, 
per the recommendation of dr. giugni to account for negative cosine similarities between words.

a function has been added to plot the frequencies of word mentions in the corpus overtime!
"""
import matplotlib.pyplot as plt
import numpy as np

def scatter(xData,yData,title,x_label,y_label):
    x = [1,2,3,4,5]
    # linear regression with least squares
    m, b = np.polyfit(x, yData, deg=1)
    plt.axline(xy1=(0, b), slope=m, label=f'$y = {m:.1f}x {b:+.1f}$')
    plt.scatter(xData,yData)
    fig = plt.gcf()
    fig.set_size_inches(5, 5)
    plt.ylim(0, 1)
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=10)
    plt.title(title,fontsize=10)
    plt.show()
```

```python
def generate_euc_dist_overtime(wordtracked, listofterms):
    #xData = ['1590-1599','1600-1609','1610-1619','1620-1629','1630-1639']
    xData = ['1590-1594', '1595-1599', '1600-1604', '1605-1609', '1610-1614', '1615-1619', '1620-1622', '1623-1626',
             '1627-1630', '1631-1635', '1636-1639']
    yData = {}
    for term in wordtracked:
        for otherterm in listofterms:
            yterm = []
            for idx,model in enumerate(moremodels): 
                if term in model.wv and otherterm in model.wv :
                    #if you would like to measure euclidean distances, this code works to do that with words as well
                    score = np.linalg.norm(model.wv[term] - model.wv[otherterm])
                else: score = 0
                yterm.append(score)
            yData[otherterm] = yterm
    font = {'fontname':'Times New Roman'}
    
    plt.figure(figsize = (16, 11))
    for term, y in yData.items():
        plt.plot(xData, y, marker = 'o')
        # plt.plot(xData, y, marker = 'o', color='pink')
        # texts.append(plt.text(xData[-1], y[-1], f'  {term}', ))
        # ax.annotate(term, (xData[-1], y[-1]), textcoords = 'offset points', xytext=(0,-15))
    plt.legend(loc='upper left', labels = yData.keys(), ncol = 2, prop={'size': 10})
    plt.title(label = wordtracked[0] + ' Word-Pairs', fontsize = 20)
    plt.xlabel('Time Period', fontsize=15)
    plt.xticks(fontsize = 14)
    plt.ylabel('Euclidean Distances', fontsize=15)
    plt.yticks(fontsize = 14)
    plt.ylim([0, 40])

generate_euc_dist_overtime(test, search)
```

```python
import os
#for tracking the frequencies of terms over time
def gettexts(folder):
    texts = []
    #list of lists of strings, each text broken up into individual token strings
    tokenized = []
    #list of texts as a continuous string
    textnames = []
    for file in os.listdir(folder):
        path = os.path.join(folder,file)
        name = os.path.basename(file)
        f = open(path,'r', encoding="utf-8")
        data = f.read()
        textnames.append(name)
        texts.append(data)
        f.close()
    for text in texts:
        #tokenize by white space
        words = text.strip().split(' ')
        tokenized.append(words)
    return [tokenized, texts, textnames]

#get the entire ep corpus, fr fr
texts1590_94 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1590-1594_texts')[0]
texts1595_99 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1595-1599_texts')[0]
texts1600_04 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1600-1604_texts')[0]
texts1605_09 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1605-1609_texts')[0]
texts1610_14 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1610-1614_texts')[0]
texts1615_19 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1615-1619_texts')[0]
texts1620_22 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1620-1622_texts')[0]
texts1623_26 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1623-1626_texts')[0]
texts1627_30 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1627-1630_texts')[0]
texts1631_35 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1631-1635_texts')[0]
texts1639_39 = gettexts('/hpc/group/datap2023ecbc/Time Series Modeling Ranges/1636-1639_texts')[0]
listoftexts = [texts1590_94, texts1595_99, texts1600_04, texts1605_09, texts1610_14, texts1615_19, texts1620_22,
               texts1623_26, texts1627_30, texts1631_35, texts1639_39] 

def freq_overtime(searchword):
    period_freq = {}
    period = 0
    for texts in listoftexts:
        count = 0
        for text in texts:
            add_to_count = text.count(searchword)
            count = count + add_to_count
        period_freq[period] = count
        period += 1
    return period_freq

def plot_freq_overtime(searchword):
    freq_dict = freq_overtime(searchword)
    xData_dict = {}
    xData = ['1590-94', '1595-99', '1600-04', '1605-09', '1610-14', '1615-19', '1620-22', '1623-26', '1627-30',
             '1631-35', '1636-39']
    index = 0
    for key, value in freq_dict.items():
        xData_dict[xData[index]] = value
        index += 1
    time_periods = xData_dict.keys()
    freqs = xData_dict.values()
    plt.figure(figsize = (16, 11))
    plt.bar(time_periods, freqs, color ='purple',
        width = 0.4)
    plt.xlabel("Time Periods", fontsize=15)
    plt.ylabel("Number of mentions in corpus", fontsize=15)
    plt.title("'" + searchword + "'" + " mentions overtime", fontsize=20)
    plt.show()
    
```
